```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from gradboost import GradBoostOnDT
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import pandas as pd 
import numpy as np
from catboost import CatBoostRegressor
```

```python
X, y = datasets.make_regression(n_samples=100, n_features=3, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
model = GradBoostOnDT(400, 0.1, 1)
model.fit(X_train, y_train)
model.train()
```

```python
predict = model.predict(X_test)
predict
```

```python
r2_score(y_test, predict)
```

```python
catboost_model = CatBoostRegressor(learning_rate=0.1, depth=1, n_estimators=400)
```

```python
catboost_model.fit(X_train, y_train)
```

```python
c_predict = catboost_model.predict(X_test)
```

```python
r2_score(y_test, c_predict)
```
